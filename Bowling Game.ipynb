```python
import random
all = []
all_rounds = {}
total = []
player_names = []
rounds = 10
total_pins = 10

num_of_players = int(input("Enter number of players: "))

   
for i in range(0, num_of_players):
    player_names.append(input("Player" + str(i + 1) + ": "))


for round in range(0, rounds):
    print("\nRound " + str(round + 1))
    player_score = []

    for player in range(0, num_of_players):
        first_roll = random.randint(0, total_pins)
        second_roll = 0
        if first_roll != 10:
            second_roll = random.randint(0, total_pins - first_roll)
        print(str(player_names[player]) + ": " + str(first_roll) + " " + str(second_roll))

        
        
        bowling=open('boling game2.txt','a')
        bowling.write(str(player_names[player]) + "\n")
        bowling.write(str(first_roll) + "\n")
        bowling.write(str(second_roll) + "\n\n")
        bowling.close()

        
        all_rounds[player] = first_roll + second_roll
        player_score.append((first_roll + second_roll))
    total.append(player_score)


total_sum_of_players = [sum(i) for i in zip(*total)]


print("\nFinal Score")
for player_name, total_score in zip(player_names, total_sum_of_players):    
    print(str(player_name) + ": " + str(total_score))


        

    








```
